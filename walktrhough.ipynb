```python
zipdf = pd.read_csv('./data/zips.csv')
dmadf = pd.read_csv('./data/data.csv')
resdf = pd.read_csv('./data/results.csv', header=None)
resdf.columns = ['postalcode', 'group']

def max_share(x):
    res = x.groupby('outletfamily_name').agg({'Dollars': np.sum})
    res['share'] = res['Dollars']/sum(res['Dollars'])
    
    return max(res['share'])

df = resdf.merge(dmadf, on='postalcode', how='inner')

groupdf = resdf.merge(dmadf, on='postalcode', how='inner').groupby('group').agg({'Dollars': 'sum', 'postalcode': 'nunique', 'outletfamily_name': 'nunique'})
groupdf['max_share'] = df.groupby('group').apply(lambda x: max_share(x))
groupdf.sort_values('max_share', ascending=False)
```
